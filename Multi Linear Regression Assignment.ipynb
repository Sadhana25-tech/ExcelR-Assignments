```python
rsq_hp = smf.ols('RD~Admin+MS+State',data=STP).fit().rsquared  
vif_hp = 1/(1-rsq_hp)

rsq_wt = smf.ols('Admin~RD+MS+State',data=STP).fit().rsquared  
vif_wt = 1/(1-rsq_wt)

rsq_vol = smf.ols('MS~RD+Admin+State',data=STP).fit().rsquared  
vif_vol = 1/(1-rsq_vol)
```

```python
rsq_sp = smf.ols('State~RD+Admin+MS',data=STP).fit().rsquared  
vif_sp = 1/(1-rsq_sp) #error as the state contains non numeric values hence , ignoring state columns
```
